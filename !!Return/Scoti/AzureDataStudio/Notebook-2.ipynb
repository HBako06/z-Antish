```sql
BEGIN TRANSACTION; -- Iniciar una transacción para asegurar la integridad de los datos

-- Bloquear la tabla para evitar operaciones concurrentes
ALTER TABLE [dbo].[JK_workersColab] SET (LOCK_ESCALATION = TABLE);

-- Cargar datos desde el archivo CSV a la tabla temporal
BULK INSERT [dbo].[Temp_JK_workersColab]
FROM 'inputs/dataDn5 lives.csv'
WITH (
    DATA_SOURCE = 'BlobDataSource2',
    --FORMAT = 'CSV', 
    FIELDTERMINATOR = ',',    -- Delimitador de campos: coma
    FIRSTROW = 2 
    --MAXERRORS = 10,
    --ROWTERMINATOR = '\n'      -- Delimitador de filas: salto de línea
);

-- Actualizar los registros existentes y/o insertar nuevos usando MERGE
MERGE INTO [dbo].[JK_workersColab] AS Target
USING [dbo].[Temp_JK_workersColab] AS Source
ON Target.[DNI] = Source.[DNI] -- DNI es tu clave primaria o tiene una restricción de unicidad
WHEN MATCHED THEN
    UPDATE SET 
        Target.[Procesado] = Source.[Procesado],
        Target.[FechaTrabajo] = Source.[FechaTrabajo],
        Target.[Status] = Source.[Status],
        Target.[IpCliente] = Source.[IpCliente] -- Actualiza las columnas necesarias
WHEN NOT MATCHED THEN
    INSERT ([DNI], [Procesado], [FechaTrabajo], [Status], [IpCliente]) 
    VALUES (Source.[DNI], Source.[Procesado], Source.[FechaTrabajo], Source.[Status], Source.[IpCliente]); -- Inserta nuevos registros

-- Limpiar la tabla temporal
TRUNCATE TABLE [dbo].[Temp_JK_workersColab];

-- Finalizar la transacción
COMMIT TRANSACTION;
```

```sql
CREATE TABLE dbo.Temp_JK_workersColab (
    DNI VARCHAR(255), -- Asegúrate de que el tipo de datos coincida con el de la columna DNI en tu tabla principal
    Procesado INT,
    FechaTrabajo DATE, -- Asegúrate de que el tipo de datos coincida con el de la columna FechaTrabajo en tu tabla principal
    Status VARCHAR(255), -- Asegúrate de que el tipo de datos coincida con el de la columna Status en tu tabla principal
    IpCliente VARCHAR(255) -- Asegúrate de que el tipo de datos coincida con el de la columna IpCliente en tu tabla principal
);
```
